```python
import numpy as np
import tensorflow as tf
```

```python
data="/content/drive/MyDrive/Covid19-dataset"
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras import layers
```

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
train_data= tf.keras.utils.image_dataset_from_directory(
    data,
    validation_split=0.32,
    subset="training",
    seed = 123,
    image_size=(img_height,img_width),
    batch_size=batch_size
)
```

```python
data_classes=train_data.class_names
```

```python
val_data = tf.keras.utils.image_dataset_from_directory(
    data,
    validation_split=0.2,
    subset="validation",
    seed = 123,
    image_size=(img_height,img_width),
    batch_size=batch_size
    
)
```

```python
no_of_classes = len(train_data.class_names)
```

```python
class_names=train_data.class_names
```

```python
model = Sequential([
 
  layers.Rescaling(1./255),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(no_of_classes, name="outputs")
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
epochs = 5
history = model.fit(
  train_data,
  validation_data=val_data,
  epochs=epochs
)
```

```python
# img_url="https://www.itnonline.com/sites/default/files/Chest.jpeg"
# img_path=tf.keras.utils.get_file("Chest_image",origin=img_url)
img_path="/content/drive/MyDrive/notcovid.jpeg"

img = tf.keras.utils.load_img(
    img_path,target_size=(img_height,img_width)
)

img_array= tf.keras.utils.img_to_array(img)
img_array=tf.expand_dims(img_array,0)


predictions = model.predict(img_array)
score= tf.nn.softmax(predictions[0])


print(f"Imgage is{class_names[np.argmax(score)]},{100*np.max(score)} ")
```
